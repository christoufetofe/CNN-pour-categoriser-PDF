```python
#Le but de ce script est de parcourrir plusieur sites web et de récupérer les liens PDF sur ces sites
```

```python
from bs4 import BeautifulSoup
import requests
import urllib.request
import time
```

```python
listeliendescraping=[]
listeliendespdf=[]
```

```python
#précharge une liste d'adresse internet pour commencer le scrapping
```

```python
for i in range (10,50,10):
    sitedepart=f'https://scholar.google.com/scholar?start={i}&q=machine+learning&hl=fr&as_sdt=0,5'
    listeliendescraping.append(sitedepart)
```

```python
#Sinon on peut utiliser les lignes en dessous pour une version plus simple et n'avoir qu'un seul site de chargé au démarrage du scraping
```

```python
#Affiche la liste des sites qui sont chargé dans la file d'attente pour le scraping
listeliendescraping
```

```python
#Les fonctions start et finish sont des compteurs et servent uniquement pour calculer le temps utiliser par l'ordinateur pour faire le scraping
#on peut les sortir
```

```python
# Le programme parcours les sites web précharger dans la file d'attente et à chaque fois qui trouve un lien il ajoute ce lien à la file d'attente
#si il trouve un fichier pdf il recépere le lien
```

```python
start = time.perf_counter()
#Changer la valeur du range permet de scraper plus ou moins de sites
#Attention avec une valeur de range trop élever le scrapeur ira surement sur un site qui a aucun rapport avec le site de départ
#Attention a ce que la valeur du range soit supérieur ou égal au nombre de sites contenus dans listeliendescraping
for i in range (40):
    link=listeliendescraping[i]
    print('i =',i)
    print(len(listeliendescraping))
    print(len(listeliendespdf))
    #print(link)

    
    
    try: 

        reponse=requests.get(link)
        soup=BeautifulSoup(reponse.text,"html.parser")
        time.sleep(1)
        recupliens=soup.find_all("a")
        #while(len(listeliendespdf)<10):
        for lien in recupliens:
            a=(lien.get('href'))
            print(a)
            if (type(a)==str):
                
                if ((a.endswith('zip')==True) or (a.endswith('rar')==True)):
                    print('un fichier a telecharger')
                else:
                    
                    if (a.startswith('http')==True):
                        if a not in listeliendescraping:
                            listeliendescraping.append(a)
                    if (a.endswith('pdf')==True):
                        if (a.startswith('/')==True):
                            b=link+a
                            if b not in listeliendespdf:
                                listeliendespdf.append(b)
                        elif (a.startswith('http')==True):
                            b=a
                            if b not in listeliendespdf:
                                listeliendespdf.append(b)
        
    except :
        pass
finish = time.perf_counter()
print(f'Finished in {round(finish-start, 2)} second(s)')
```
